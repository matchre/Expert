# COLLECTION VIGNETTES PANINI D'UNE EQUIPE DE 11 JOUEURS

```python

from sympy import *
A=Matrix([[0 for i in range(12)] for j in range(12)])
for i in range(11):
    A[i,i+1]=1-i/11
    A[i,i]=i/11
A[11,11]=1
A.evalf(n=2)
```

```python
#au départ
P=zeros(1,12)
P[0,0]=1
P.evalf(n=2)
```

```python
#au bout de 10 CARTES ACHETees
for k in range(10):
    P=P*A
P.evalf(n=2)


```

```python
# PROB D'avoir au moins n cartes avec un seuil donnée
def combien(seuil=0.70,n=9):
    P=zeros(1,12)
    P[0,0]=1
    compt=0
    while sum(P[0,n:])<seuil:
        P=P*A
        compt+=1
    return(compt)
combien(0.70,9)
combien(0.75,9)""
```

```python
for seuil in range(70,100,5):
    print("POUR AVOIR AU MOINS 9 CARTES AVEC ",seuil," % DE CHANCES, IL FAUT ACHETER AU MOINS ",combien(seuil/100,9))
```

```python
for seuil in range(70,100,5):
    print("POUR AVOIR  11 CARTES AVEC ",seuil," % DE CHANCES, IL FAUT ACHETER AU MOINS ",combien(seuil/100,11))
```
