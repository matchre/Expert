##  1er combat:  "FPP" VS "PPF" 

```python
# matrice de transition de FFF,FFP;FPF,FPP,PFF,PFP,PPF,PPP
from sympy import *
A=Matrix([[0 for i in range(8)] for j in range(8)])
A[0,0]=0.5
A[0,1]=0.5
A[1,2]=0.5
A[1,3]=0.5
A[2,4]=0.5
A[2,5]=0.5
A[3,3]=1# une fois en FPP on y reste
A[4,0]=0.5
A[4,1]=0.5
A[5,2]=0.5
A[5,3]=0.5
A[6,6]=1# une fois en PPF on y reste
A[7,6]=0.5
A[7,7]=0.5
A.evalf(n=1)
```

```python
#au départ, on a autant de chances de choisir n'importe quelle triplette
P=1/8*ones(1,8)
P.evalf(n=3)
```

```python
#au bout de 100 PIECES LANCEES
for k in range(100):
    P=P*A
P.evalf(n=2)
```

```python
from sympy import *
def Result(Alice,Bob):
    a=0
    b=0
    for i in range(3):
        if Alice[i]=="P":
            a=a+pow(2,2-i)
        if Bob[i]=="P":
            b=b+pow(2,2-i)
    A=Matrix([[0 for i in range(8)] for j in range(8)])
    A[0,0]=0.5
    A[0,1]=0.5
    A[1,2]=0.5
    A[1,3]=0.5
    A[2,4]=0.5
    A[2,5]=0.5
    A[3,6]=0.5 
    A[3,7]=0.5
    A[4,0]=0.5
    A[4,1]=0.5
    A[5,2]=0.5
    A[5,3]=0.5
    A[6,4]=0.5 
    A[6,5]=0.5
    A[7,6]=0.5
    A[7,7]=0.5
    A.evalf(n=1)
    A[a,:]=Matrix([[0 for i in range(8)]])
    A[a,a]=1
    A[b,:]=Matrix([[0 for i in range(8)]])
    A[b,b]=1
    P=1/8*ones(1,8)
    for k in range(100):
        P=P*A
    return(P[a].evalf(n=3), P[b].evalf(n=3))
Result("FPF","FFP")
    
```
