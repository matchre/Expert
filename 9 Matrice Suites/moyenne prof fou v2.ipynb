# PROF FOU !!!
 Il va vous noter d'une façon peu recommandable! 
 
 Il décide de commencer par mettre 20 à tous les élèves de votre classe.
 
 Puis, chaque jour, il choisit au hasard un élève de votre classe pour modifier sa note : 
 
### s'il a 20, il la remplace par 0; et s'il a 0, il la transforme à nouveau en 20. 
 
 Au bout de 120 jours, ce « professeur » écrit sur le bulletin la dernière note qu'il a attribuée à chaque élève. 
 
 La moyenne de la classe conviendra-t-elle à votre Proviseur? 
 

```python
# construction de la matrice de passage 
from sympy import *
N=17 # nbre d'eleve
A=Matrix([[0 for i in range(N+1)] for j in range(N+1)])
for i in range(N):
    A[i,i+1]=1-i/N
    A[i,i-1]=i/N

A[N,N-1]=1
A.evalf(n=2)
```

```python
# ou sa vrsion exact
from sympy import *
N=17
A=Matrix([[0 for i in range(N+1)] for j in range(N+1)])
for i in range(N):
    A[i,i+1]=Rational(N-i,N)
    A[i,i-1]=Rational(i,N)

A[N,N-1]=1
A
```

```python
#au départ, tous ont 20 !
# P[0,K] DONNE LA PROB D AVOIR K  ELEVES QUI ONT ZERO DANS LA CLASSE 
P=zeros(1,N+1)
P[0,0]=1
P.evalf(n=2)
```

```python
au bout d'un jour un élève a forcément zéro
```

```python
def auboutdeNBjours(NB):
    P=zeros(1,N+1)
    P[0,0]=1
    for i in range(NB):
        P=P*A
    return(P)
auboutdeNBjours(1) # P[0,1]=1 ON EST SUR QU'UN ELEVE AIT ZERO
```

au bout de 120 JOURS, la répartition théorique vaut:

```python
auboutdeNBjours(120).evalf(n=2)
```

```python
auboutdeNBjours(120)[0,6].evalf(n=2)
```

il y a par exemple 19% de chances que 6 élèves un zéro, la moyenne de la classe vaudrait alors:

```python
round((6*0+(N-6)*20)/N,2)
```

on peut calculer les moyennnes dans chachun des cas probables :

```python
moyenne=0
for i in range(N+1):
    if auboutdeNBjours(120)[0,i].evalf(n=2)>0:
        print("il y a par exemple ",100*auboutdeNBjours(120)[0,i].evalf(n=2)," % de chances que ",i," élèves aient un zéro")
        print(" la moyenne de la classe vaudrait alors ",round((i*0+(N-i)*20)/N,2))
        moyenne=moyenne+auboutdeNBjours(120)[0,i].evalf(n=2)*round((i*0+(N-i)*20)/N,2)
            
```

on peut ainsi **espérer** après de tels agissements de ce prof fou d'avoir une moyenne

```python
moyenne
```

une simulation:au bout de 120 jours, où en est- on ? répartition des notes et moyennes de la classe

```python
def apres120jour():
    ElevesQuiA20=[]
    #tous éléves ont 20 au départ
    for j in range(N):
        ElevesQuiA20.append(20)
    #Choisir un élève et lui changer sa note de 20 à à et réciproquement.
    for i in range(120):
        numero = randint(0, N - 1)
        if ElevesQuiA20[numero] == 20:
            ElevesQuiA20[numero] = 0
        else:
            ElevesQuiA20[numero] = 20
    return ElevesQuiA20,round(sum(ElevesQuiA20)/N,2)
apres120jour()
```

une autre simulation:au bout de 120 jours, où en est- on ?

```python
apres120jour()
```

pour lisser la fluctuation, on va faire 10 000 simulations et faire la moyenne de ces moyennes

```python
moy=0
for j in range(10000):
    moy=moy+apres120jour()[1]
print(moy/10000)
    
```
