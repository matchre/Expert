# Cryptage Secret à 3
Le partage de clé secrète de Shamir est un algorithme de cryptographie. 

C'est une forme de partage de secret, où un secret est divisé en parties, donnant à chaque participant sa propre clé partagée, où certaines des parties ou l'ensemble d'entre elles sont nécessaires afin de reconstruire le secret.


Pour le secret à 3, on se base sur des fonctions et coordonnées mathématiques. Comme son nom l’indique, il nous faut donc 3 points donnés à 3 personnes pour retrouver une fonction polynôme (de forme ax²+bx+c): si on ne donne que 2 points on ne peut pas déduire la fonction passant par les points car on aurait une droite, il existerait donc une infinité de parabole passant par les 2 points.

![](https://codi-lille.beta.education.fr/uploads/upload_3f659a2f166bad068e0b9306bec98fe5.png)

Il nous faut donc impérativement 3 personnes (ou plus pour Hugo) pour résoudre un système, une personne à un seul et même point qui lui est attribué: (xi;yi) on a donc:

( x1 ; y1 ) → Personne 1

( x2 ; y2 ) → Personne 2

( x3 ; y3 ) → Personne 3


```python
#coding: UTF-8


import numpy as np
import random #on entre les variables a,b,c
a = int(input("Entrer la première variable de la fonction",))
b = int(input("Entrer la deuxième variable de la fonction",))
c = int(input("Entrer la troisième variable de la fonction",))

x1=random.randint(-100, 100) #x reçoit une valeur entre -100 et 100
x2=random.randint(-100, 100)
while x2==x1:    #on attribue à x2 une valeur différente de x1
    x2=random.randint(-100, 100)
x3=random.randint(-100, 100)
while x3==x1 or x3==x2 : #on attribue à x3 une valeur diférente de x2 ET de x1
    x3=random.randint(-100, 100)

def f(x):
    return a*x*x+b*x+c #définition de la fonction

y1=f(x1) #la personne 1 reçoit une valeur par la fonction où x est un chiffre aléatoire
y2=f(x2)
y3=f(x3)

X = np.array([[x1*x1,x1,1],[x2*x2,x2,1],[x3*x3,x3,1]]) #attribution
Y = np.array([y1,y2,y3]) #variable à retrouver
Coeff = np.linalg.solve(X, Y)  #calcul de la matrice

print ("La première personne reçoit le couple","(",x1,";",y1,")",sep='')
print ("La deuxième personne reçoit le couple","(",x2,";",y2,")",sep='')
print ("La troisième personne reçoit le couple","(",x3,";",y3,")",sep='')
```

```python
#coding: UTF-8
from sympy import * #Les fonctions symbols pour faire du calcul littéral et simplify pour simplifier (légèrement) le polynôme résultat.
from numpy import *

def Lagrange(Lx, Ly):
    X = symbols('X')
    if  len(Lx) != len(Ly): #Si on ne donne pas le même nombre d'ordonnées et d'abscisses
        print("Les listes font pas la même longueur, c'est génant non ?")
        return 1
    d = 0
    for k in range(len(Lx)):
        t = 1
        for j in range(len(Lx)):
            if j != k:
                t = t * ((X - Lx[j]) / (Lx[k] - Lx[j]))     # Calcul des polynômes de Lagrange
        d += t * Ly[k]          #Ajout des polynômes de Lagrange au polynôme principal
        d = simplify(d)         #Simplification (légère) d'écriture
    return d

Listex = []
Listey = []
x, y = 0, 0
compteur = 1

while x != '': #Demande des abscisses et ordonnées des points
    x = input('''Entrez une abscisse s'il vous plaît''')
    if x != '':
        Listex += [int(x)]
    if x == '':
        break
    y = input('''Entrez une ordonnée s'il vous plaît''')
    if y != '':
        Listey += [int(y)]


print(Lagrange(Listex, Listey))

```
